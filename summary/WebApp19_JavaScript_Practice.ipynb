* input 태그의 id속성과 JS에서 제공하는 API인 getElementById 내장함수 활용
* getElementById는 document 객체의 메소드(함수)임.

    그 객체의 value라는 속성에 텍스트 창에 사용자가 입력한 값이 저장됨.
* 주의: JS는 객체지향언어이며, 많은 값들이 객체이다.
    따라서 속성과 메소드(함수)의 개념이 기본적으로 사용된다.

* document.getElementById의 리턴값 객체의 className 속성을 지정하는 방식을 활용한다.
```html
<!DOCTYPE html>
<html>
<head>
     <meta charset="utf-8">
   <style>
     .em{
       text-decoration:underline;
     }
   </style>
</head>
<body>
  <ol id="target">
    <li>html</li>
    <li>css</li>
    <li>javascript</li>
  </ol>
  <ul>
    <li>최진혁</li>
    <li>최유빈</li>
    <li>한이람</li>
    <li>한이은</li>
  </ul>
  <input type="button" value="강조" onclick="document.getElementById('target').className='em'" />
</body>
</html>
```

## MVC 패턴 소개

* MVC: Model-View-Controller(모델-뷰-컨트롤러)의 약자
* MVC 패턴은 프로그램을 말 그대로 모델, 뷰, 컨트롤러의 세 요소로 구성되도록 구현하는 방식을 의미함.
* 모델(Model): 주시 대상
    * 위 예제에서: white 버튼과 black 버튼이 주시 대상이다.
* 뷰(View): 주시 대상 자체와 주시 대상이 작동하는 방식을 보여주는 사용자 인터페이스(UI)
    * 위 예제에서: 버튼을 담고 있는 웹페이지 전체
* 컨트롤러(Controller): 모델과 뷰를 연결하는 역할 수행, 즉, 모델을 조작해서 뷰를 통해 사용자에게 보여주도록 함.
    * 위 예제에서: 땡땡땡(`...`) 부분에 정의된 JS 코드
* 아래 그림은 MVC 패턴이 작동하는 방식을 보여준다. 
    * 참조 사이트: https://opentutorials.org/course/697/3828
