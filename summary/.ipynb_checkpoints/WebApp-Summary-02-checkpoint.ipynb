## 의미가 잘 드러나는 웹 (Semantic Web)

* 사람뿐만 아니라 웹브라우저 등의 기계에도 정보를 보다 더 잘 전달하도록 하는 태그들을 활용함.
* CSS를 함께 사용할 때 중요 역할 수행
* 다른 특별한 기능은 없음. 즉, 내용에 어떤 변화를 주지 않음.
* 예제: `nav` 태그
    * 탐색용 부분이라는 정보를 웹브라우 알려주는 태그
    * nav = navigation

</p>

## index.html 파일

* 웹사이트의 대문 페이지 역할을 수행하는 특수한 파일
* 특정 사이트의 메인 페이지를 방문할 경우 서버는 우선적으로 `index.html` 파일 내용을 전달해줌.
* 대문 역할에 해당하는 내용만 남겨 두고 다른 내용은 내용 별로 분리하여 여러 개의 html 파일을 생성함.
* 대신에 `index.html` 파일에 다른 페이지에 해당하는 **링크**를 작성해야 함.

<p>
<table cellspacing="20">

<tr>
<td>
<img src="images/internet-web27.png", width=300>
</td>
</tr>

</table>
</p>

* 앞서 배운 태그들을 활용하여 각각의 페이지의 내용을 담은 html 파일을 생성해야 함.
    그렇지 않으면 링크를 눌렀을 때 오류 발생.
    
* 위 그림에 포함된 코드의 경우 아래 세 개의 파일이 추가로 생성되야 함.
    * `page_html.html`: JavaScriptf란 주제 관련 내용
    * `page_vc.html`: 변수와 상수 관련 내용
    * `page_op.html`: 연산자 관련 내용
 
<p>
<table cellspacing="20"> 
<tr>
<td>
<img src="images/internet-web28.png", width=200>
</td>
<td>
<img src="images/internet-web29.png", width=200>
</td>
<td>
<img src="images/internet-web30.png", width=200>
</td>
</tr>

</table>
</p>

* header에도 링크를 주고 완성하면 다음과 같다.

<p>
<table cellspacing="20">

<tr>
<td>
<img src="images/internet-web31.png", width=300>
</td>
</tr>

</table>
</p>

* **주의사항:** 
    * 이렇게 하면 다른 파일에서도 header 부분을 수정해야 불편함이 발생한다. 
        몇 개 정도는 가능하지만 수 백, 수 천 개의 페이지를 수정하는 일은 불가능하다.
        따라서 다른 기술이 요구된다.

    * 앞으로 CSS, Javascript, php 언어와 코딩 디자인 요령 등을 터득하면서 보다 효율적이면서
        멋있는 웹디자인을 구현하는 방법을 터득하게 될 것이다.
