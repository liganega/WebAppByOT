# 웹브라우저와 웹서버의 관계 정리

* 웹브라우저에서 웹페이지 또는 파일 주소를 이용해 웹서버에 html, css, js 등의 특정 파일을 요청
* 웹서버는 요청받은 파일을 웹브라우저에게 건네 줌
* 웹브라우저는 건네 받은 파일의 내용을 적절한 방식으로 화면에 보여줌.

<p>

* 반면에 php와 데이터베이스는 서버쪽에서 실행된다.
    * 웹브라우저가 php 파일을 요청할 때 웹서버는 단순히 해당 파일을 건네 주지 않는다.
    * 대신에 먼저 php 태그로 감싸인 코드를 php 엔진으로 실행한 결과를 html로 
        변환하여 변경된 html 파일을 웹브라우저에게 전달한다.

# Javascript 대 PHP

* 서로 경쟁적이며 협력하는 관계
* 많은 동일한 일을 양쪽 모두에서 할 수 있다.
* 반면에 각각의 장점이 있다.

## javascript와 php의 특징

* html와 css는 정적인 언어이다. 
    * 정적: 한 번 만들면 내용의 변화 없음.
    * 문서 만들기에 좋음    
* javascript와 php 동적인 언어이다.
    * 상황에 따라 다르게 기능하는 애플리케이션 만들 때 활용.

### javascript

* 예를 들어, black 또는 white 버튼을 누르면 바탕화면의 색깔을 변경하도록 만들 수 있다.

* html 에서 css 파일을 따로 분리하여 스타일 파일을 여러 파일에서 공유하는 것처럼
    하나의 html 파일을 여러 html 파일이 공유할 수 있는 기능 제공

* 예제: `nav` 태그로 감싸인 네비게이션 부분을 수정하고 할 때 모든 파일을 수정해야만 한다.
    예를 들어, `ol` 태그를 `ul` 태그로 변경하고자 하면 다른 파일에서도 수정해야 한다.

# CSS 이론

* CSS = Cascading Style Sheet
* 스타일을 꾸며주는 언어
* html과는 완전히 다른 언어이지만, html과는 별개로 사용될 수 없음
* `head` 태그 안에서 `style` 태그로 감싸임

## CSS 기초

* 예제: h1 태그의 색상을 빨강으로 바꿀 때

<p>

* 위 예제에서 `h1`은 **선택자**라 불리고 나머지 부분은 **서술**이라고 불린다.

<p>

* 서술에는 `color` 속성 이외에 다양한 속성들이 있다.
    * `font-size`: 글자 크기. 픽셀(px) 단위로 설정
    * `text-decoration`: 밑줄 긋기 등 설정
    * `border`: 상자로 감싸기 옵션
    * 등등
    * 속성은 세미콜론(`;`)으로 구분됨.
    
* 선택자는 여러 개를 동시에 또는 따로 따로 지정할 수 있다.
    * 동시 선택: h1, h2 { 속성들 }
        * 쉼표에 주의

* 중첩 선택자 사용 가능
    * 예를 들어, header 태그 안의 h1 태그에만 속성 정할 때: header h1 { 속성들 }
        * 쉼표 없음
