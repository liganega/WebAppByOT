# 인터넷과 웹의 차이

* 인터넷: 
    * 1960년대 등장
    * 전 세계 컴퓨터들을 연결하는 네트워크
    * 자료 공유 가능
    * 인터넷 상에 웹, ftp, 이메일 등이 돌아감
    
* 웹: 컴퓨터가 html로 작성된 정보를 주고 받도록 고안된 시스템
    * 1990년대 등장
    * 창시자: 팀 버너스리
        * 웹브라우저와 서버 모두 개발

# 서버와 클라이언트

서버와 클라이언트는 컴퓨터로서 동일하다. 
다만 안정성과 속도의 차이가 있을 수 있다.

* 클라이언트
    * 브라우저에 웹사이트 주소 입력 하면, 웹서버에 웹페이지 문서 요청
    * 웹서버가 응답한 웹페이지를 웹브라우저를 통해 보여줌
    
* 서버
    * 클라이언트의 요청에 응답하기 위해 필요한 기능과 자료를 저장함
    
인터넷 상에 존재하는 게임, 모바일 앱 등은 클라이언트와 서버의 관계를 활용하여 동작한다.

# HTML 기본문법

* 서버와 클라이언트는 웹브라우저를 통해 웹서버에 특정 웹페이지 자료를 요청함
* 웹서버는 html 파일을 전달하는 방식으로 응답함
* 웹브라우저는 전달 받은 html 파일 내용을 보여줌.

* HTML = HyperText Markup Language

        * 링크를 누르면 다른 문서로 연결됨
        * 링크가 HTML의 본질, 따라서 웹의 본질임.

**예제:  strong 태그**

* 설명: 강조을 위한 태그
* 코드 예제: ex_html.html

```html
<!DOCTYPE html>
<html>
  <head>
    <meta charset="utf-8" />
  </head>
  <body>
    안녕하세요. <strong>생활코딩</strong>입니다.
  </body>
</html>
```

**예제: `a` 태그**

* 설명: 링크를 걸기 위한 태그
* 다양한 속성을 가짐
    * `href`: 링크로 연결된 주소를 가리킴
        * 속성값: 웹페이지 주소
    * `target`: 링크로 연결된 웹페이지를 독립된 새 창에서 열고자할 때 활용
        * 속성값: `_blank` 또는 `_self` 등
    * 속성과 속성은 공백으로 구분된다.

<p>
<table cellspacing="20">

<tr>
<td>
<img src="images/internet-web09.png", width=300>
</td>

</tr>

</table>
</p>
    
* 코드 예제: ex_html2.html

```html
<!DOCTYPE html>
<html>
  <head>
    <meta charset="utf-8" />
  </head>
  <body>
    안녕하세요. <a href="http://opentutorials.org/course/1" target="_blank">생활코딩</a>입니다.
  </body>
</html>
```

### `html` 태그

* html언어 문서임을 보여주는 태그
* `head`와 `body` 태그 전체를 깜싼다.
    
<p>
<table cellspacing="20">

<tr>
<td>
<img src="images/internet-web17.png", width=300>
</td>

</table>

* 코드 예제: ex_html3.html

```html
<!DOCTYPE html>
<html>
  <head>
    <meta charset="utf-8" />
    <title>생활코딩</title>
  </head>
  <body>
    <ol>
      <li>html</li>
      <li>css</li>
      <li>JavaScript</li>
    </ol>

    <ul>
      <li>최진혁</li>
      <li>최유빈</li>
      <li>한이람</li>
    </ul>
  </body>
</html>
```

# HTML 버전 정보

* 독타입(`DOCTYPE`)은 사용하는 html 언어의 표준에 대한 정보를 전달한다. 
    어떤 표준에 따라 작성된 파일인지 웹브라우저가 알 수 있도록 도와준다.

<p>

</p>

* 주의: 독타입 정보는 html 바깥에 위치한다. 
    굳이 없어도 작동은 하지만, 보다 정확하게 사용된 언어에 대한 정보를 브라우저에게 전달하기 위해 필요하다. 
    이유는 여러 표준안이 존재하며 발전해왔으며, 표준안에 따라 태그들이 달라지기 때문이며, 
    현재는 HTML 5 표준안을 따른다고 적혀 있음.
