## 실습 2

index.html 페이지와 동일하게 다른 html 페이지도 수정하고자 함.
하지만, 여러 개의 파일을 일일이 수정하는 일은 매우 고되고 때로는 불가능함.
따라서 css 파일을 따로 분리, 저장해서 활용함.

### style.css 파일 생성

예를 들어, style.css 파일에 스타일 부분 내용을 복사해서 옮김. 
그리고 원래 있던 자리에서는 아래 내용을 첨부함.
    
```
<link rel="stylesheet" type="text/css" href="http://localhost/style.css">
```

<p>
<table cellspacing="20">
<tr>
<td>
<img src="images/internet-web48.png", width=500>
</td>
</tr>
</table>
</p>

<p>
<table>
<tr>
<td>
<img src="images/internet-web49.png", width=500>
</td>
</tr>
</table>
</p>    

* index.html 파일 뿐만 아니라 다른 파일에도 동일하게 적용하면 됨.

* 이제 스타일 파일의 내용만 변경하면 모든 사이트에 동일한 스타일이 적용된다.

## 디자인 패턴 원리: 중복 제거

많은 웹사이트를 관리해야 할 경우, 중복된 CSS 스타일 파일을 독립시켜서 활용하면
코드의 중복을 제거할 수 있다.

* 중복제거 이전

<p>
<table cellspacing="20">
<tr>
<td>
<img src="images/internet-web53.png", width=500>
</td>
</tr>
</table>
</p>

* 중복제거 이후

<p>
<table>
<tr>
<td>
<img src="images/internet-web54.png", width=500>
</td>
</tr>
</table>
</p>    

* 장점:
    * 관리의 용이성이 증가한다.
    * 서버에 주는 부담, 다운로드 메로리 사용량, 실행 시간 등을 많이 줄인다.
