# 웹브라우저와 웹서버의 관계 정리

* 웹브라우저에서 웹페이지 또는 파일 주소를 이용해 웹서버에 html, css, js 등의 특정 파일을 요청
* 웹서버는 요청받은 파일을 웹브라우저에게 건네 줌
* 웹브라우저는 건네 받은 파일의 내용을 적절한 방식으로 화면에 보여줌.

### 문자열

* 문자열은 따옴표로 감싸진다.
    * `"10"`은 숫자 10이 아니라 문자열이다.
